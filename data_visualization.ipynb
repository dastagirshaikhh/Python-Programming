```python
from itertools import groupby
import pandas as pd
import os
# Setting the working directory
os.chdir("C:\\Users\\dasta\\OneDrive\\Documents\\Data Science Tabular sheets")
df = pd.read_csv("Car_sales.csv")  # Displaying the data

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df
```

```python
for group, data in df.groupby('Manufacturer'):
    print (group, data['4-year resale value'].mean()) #printing the mean value of a column by grouping columns
```

```python
#Displaying the data in the form of Horizontal Bar Graph
plt.figure(figsize=(25,8))
for group, data in df.groupby('Manufacturer'):
    plt.barh(group, data['4-year resale value'].mean()) #Displays the data in pie chart using matplotlib
    plt.xlabel ("4 Year resale value Average")
    plt.ylabel ("Manufacturer")
    plt.title ("4 Year resale Average value vs Manufacturer Graph")
```

```python
sns.boxplot(data = df, y = "Vehicle type", x = "Fuel efficiency") #box plot

#What is the use of line in the box?
# The line in the box plot works as a graph suppose there is a class of 65 students and each of them are owns a laptop
# 52 students owns laptops in the range of 30k to 70k and 8 students own laptops above 70k. So the median of 65 students
# is 33 if the line in the box follows the median and if the value in the right side of the median is higher than the 
# line in the box will appear on the right side and if the values are higher on the left side then the line in the box
# will appear on the left side.
```
